```python
from sklearn.datasets import load_digits
```

```python
data = load_digits()
```

```python
X = data.data
y = data.target
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
estimator = DecisionTreeClassifier()
```

```python
val = cross_val_score(estimator, X, y, cv =10)
```

```python
print(val.mean())
```

```python
with open ('result_hw_4_week.txt', 'w') as file:
    file.write(str(val.mean()))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bagging_estimator = BaggingClassifier(DecisionTreeClassifier() ,n_estimators= 100)
res_2 = cross_val_score(bagging_estimator, X, y, cv = 10)
```

```python
with open('result_hw_4_week_2.txt', 'w') as file:
    file.write(str(res_2.mean()))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import matplotlib.pyplot as plt
%pylab inline
```

```python
figure = plt.figure(figsize = (10,6))
depth_list = range(1,100,5)
oy_to_plot = []
for depth in depth_list:
    estimator = RandomForestClassifier(max_depth=depth)
    quality = cross_val_score(estimator, X, y, cv = 10)
    oy_to_plot.append(quality.mean())
plt.plot(depth_list, oy_to_plot)
```

```python
figure_2 = plt.figure(figsize = (16, 10))
features = range(1, 100, 5)
oy_to_plot_2 = []
for feature in features:
    estimator = RandomForestClassifier(max_features=feature)
    quality = cross_val_score(estimator, X, y, cv = 10)
    oy_to_plot_2.append(quality.mean())
plt.plot(features, oy_to_plot_2)
```

```python
figure_3 = plt.figure(figsize = (16,10))
number_of_trees = range(1,100,5)
oy_to_plot_3 = []
for number in number_of_trees:
    estimator = RandomForestClassifier(n_estimators=number)
    quality = cross_val_score(estimator, X, y, cv = 10)
    oy_to_plot_3.append(quality.mean())
plt.plot(number_of_trees, oy_to_plot_3)
```
